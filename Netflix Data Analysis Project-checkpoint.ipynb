While exploring the data, I noticed a problem with the 'productin_countries' and 'genres' column: They both contain multiple values instead of single values, so I created a function to randomly pull out one country and one genre where exist many

```python
# created function to randomly choose value from a Dataframe's column if ceryain conditions exist

def repair_array_bound_categories(arr):
    arr = ast.literal_eval(arr)
    
    if len(arr) == 0:
        return np.nan
    
    elif len(arr) == 1:
        return arr[0]
    
    else:
        return random.choice(arr)
```

```python
# applying the above function to the Dataframe object

df["production_countries"] = df["production_countries"].apply(repair_array_bound_categories)
df['genres'] = df['genres'].apply(repair_array_bound_categories)

df.head(3)
```

After the application of different data cleaning techniques, the data is now clean and set for visualization and further analysis

# Figure_3

```python
# For the purpose of this analysis, I subset and sliced the dataset from year 2000 to year 2022. 
# Reason being that i want the plot to depict trend of release of content not backward than year 2000. 

year = df.groupby(['release_year', 'type']).size().reset_index().rename(columns = {0:"counts",'release_year':'Year'})
mod_year = year[year["Year"] >= 2000]
mod_year.head()
```

```python
chart = sns.relplot(x="Year", y="counts", data=mod_year, kind="line", hue="type", markers=True)
chart.fig.suptitle("Netflix Movies & Tv Shows Release by Year", y=1.03)
chart.set(xlabel="Year of Release")

plt.show()
```

### Answer to Question 3

Netflix is an inspirational example of a company that successfully shifted their business model multiple times and grew exponentially because of that. It has created more than 1,500 original series. In 2021 alone it released over 150 original movies but there was a sharp decline in it yearly release of content as noticeable in the chart between 2019 and 2020. Among other factors, the break out of covid 19 pandemic in that period greatly affected businesses globally.

# Figure_5

```python
# plot for the Average Runtime by Genres
genre_avtime = df.groupby("genres")["runtime"].mean().sort_values(ascending = False).reset_index().rename(columns = {0:"avg_time"})
genre_avtime.plot(x="genres", y="runtime", kind="bar", color="indigo")
plt.title("Average Runtime by Genres")
plt.xlabel("Genres")
plt.ylabel("Time (Minutes)")
plt.show()
```

### Answer to Question 5
